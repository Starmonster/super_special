```python
order=False
df_read=False
sell_open=29112
buy_open=29235
tp=float()
stop=float()
while True:
    if datetime.datetime.now().second%2==0 and df_read==False and order==False:
        try:
            df = pd.read_csv("prices_test.csv")
#             if df.shape[0]>1:
#                 print("we have a df")
        except:
            print("we're passing")
            pass
        
        # OPEN ORDER STATEMENTS
        if df.iloc[-1].low<=sell_open:
            trade.create_market_order('XBTUSDM', 'sell', '5', size=100)
            order="sell"
            tp = sell_open*0.9942
            stop = sell_open*1.005
            print(f"sell executed tp:{tp}, stop:{stop}")
        elif df.iloc[-1].high>=buy_open:
            trade.create_market_order('XBTUSDM', 'buy', '5', size=100)
            order="buy"
            tp = buy_open*1.0058
            stop = buy_open*0.995
            print(f"buy executed tp:{tp}, stop:{stop}")
        df_read=True
    
        
    elif datetime.datetime.now().second%2==0 and df_read==False and order!=False:
        try:
            df = pd.read_csv("prices_test.csv")
#             if df.shape[0]>1:
#                 print("we have a df")
        except:
            print("we're passing")
            pass
        
        if df.iloc[-1].low<=tp and order=="sell":
            ## Close sell order in profit
            trade.create_market_order('XBTUSDM', 'sell', '5', closeOrder=True)
            # Reset the sell open to a much lower price
            sell_open = sell_open*0.9
            order=False
        
        elif df.iloc[-1].high>=tp and order=="buy":
            ## Close buy order in profit
            trade.create_market_order('XBTUSDM', 'buy', '5', closeOrder=True)
            # Reset the buy open to a much higher price
            buy_open = buy_open*1.1
            order=False
            
        elif df.iloc[-1].high>=stop and order=="sell":
            ## Stop out sell order
            trade.create_market_order('XBTUSDM', 'sell', '5', closeOrder=True)
            order=False
        
        elif df.iloc[-1].low<=stop and order=="buy":
            ## Stop out buy order
            trade.create_market_order('XBTUSDM', 'buy', '5', closeOrder=True)
            order=False
        df_read=True
            
    elif datetime.datetime.now().second%2!=0 and df_read==True:
        df_read=False
        print("df switch pressed")
        
        
```
